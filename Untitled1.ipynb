```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('medical_examination.csv')
df['overweight'] = (df['weight'] / ((df['height']/100 ) **2)).apply(lambda x: 1 if x>25 else 0)
df['cholesterol'] = df['cholesterol'].apply(lambda x: 1 if x >1 else 0  )
df['gluc'] = df['gluc'].apply(lambda x: 1 if x >1 else 0  )
df
```

```python
cat = pd.melt(df, id_vars =['cardio'], value_vars = ['cholesterol', 'gluc', 'smoke' , 'alco', 'active', 'overweight'])
cat = cat.groupby(['cardio', 'variable', 'value'],as_index=False)['value'].size().reset_index().rename(columns={0:'total'})
cat
```

```python
sns.catplot( x = 'variable', y='total', hue='value', col='cardio',
                data=cat,
                kind="bar")
```

```python
heat = df[(df['ap_lo'] <= df['ap_hi']) & (df['height'] >= df['height'].quantile(0.025)) & (df['height'] <= df['height'].quantile(0.975)) & (df['weight'] >= df['weight'].quantile(0.025)) & (df['weight'] <= df['weight'].quantile(0.975))]
corr = heat.corr()
mask = np.triu(np.ones_like(corr))
fig, ax = plt.subplots(figsize=(7, 5))

    # Draw the heatmap with 'sns.heatmap()'
ax = sns.heatmap(corr.round(1),fmt= '.1f', mask=mask, vmax=.3,square=True, annot=True)
ax
```
